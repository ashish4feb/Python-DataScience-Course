```python
import pandas as pd
df = pd.read_csv('italy_earthquake.csv')
```

```python
import matplotlib.pyplot as plt
#Below line help you plot without using show() function.
%matplotlib inline
```

```python
#To plot line graph against Date we can make Date as an Index first.
df.set_index('Date',inplace=True)
```

```python

#Start a figure
plt.figure(1,figsize = (20,10))      

#define which subplot you want to plot
plt.subplot(111)
#plot the sub graph
plt.plot(df[(df.Latitude<42.800) & (df.Longitude<13.00)]['Magnitude'])

#Now if we do not specify which subplot we want to work in it would take the previous subplot for each new plot
plt.plot(df[(df.Latitude<42.500) & (df.Longitude<13.100)]['Magnitude'])
plt.plot(df[(df.Latitude>43.100) & (df.Longitude>13.100)]['Magnitude'])

#define the legend as required
plt.legend(['Latitude<42.800 & Longitude<13.00' , 'Latitude<42.500 & Longitude<13.100' , 'Latitude>43.100 & .Longitude>13.100'])
```
